```python
import pandas as pd
df = pd.read_csv('Students Social Media Addiction.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop('Student_ID', axis=1)
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['Addicted_Score'], bins=10, kde=True)
plt.title('Distribution of Addiction Scores')
plt.xlabel('Addicted Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Addicted_Score', data=df)
plt.title('Usage Hours vs Addiction Score')
plt.xlabel('Average Daily Usage (Hours)')
plt.ylabel('Addicted Score')
plt.show()
```

```python
sns.scatterplot(x='Sleep_Hours_Per_Night', y='Addicted_Score', data=df)
plt.title('Sleep Hours vs Addiction Score')
plt.xlabel('Sleep Hours Per Night')
plt.ylabel('Addicted Score')
plt.show()
```

```python
sns.scatterplot(x='Mental_Health_Score', y='Addicted_Score', data=df)
plt.title('Mental Health Score vs Addiction Score')
plt.xlabel('Mental Health Score')
plt.ylabel('Addicted Score')
plt.show()
```

```python
sns.boxplot(x='Most_Used_Platform', y='Addicted_Score', data=df)
plt.title('Platform vs Addiction Score')
plt.xlabel('Platform')
plt.ylabel('Addicted Score')
plt.xticks(rotation=45)
plt.show()
```
